```python
import matplotlib.pyplot as plt
import os
import numpy as np

T =np.array([],dtype=np.double)
P =np.array([],dtype=np.double)
E =np.array([],dtype=np.double)
p2 =np.array([],dtype=np.double)

with open("stats_var.txt","r") as f:
    while True:
        a = f.readline()
        a=a.split()
        if a == []:
            break

        T = np.append(T,np.double(a[0]))
        P = np.append(P,np.double(a[1]))
        E = np.append(E,np.double(a[2]))
        p2 = np.append(p2,np.double(a[3]))
print("temperture is :",T)
print("\n")

print("P is :",P,"\n")
print("E is :",E,"\n")
print("P2 is :",p2,"\n")   
```

```python
def plots(x,y,title):
    plt.style.use('dark_background')
    fig, ax = plt.subplots()

    x =np.array(T)
    y =np.array(y)
    ax.plot(x,y)
    ax.set_ylim(np.amin(y),np.amax(y))
    ax.set_xlim(np.amin(x),np.amax(x))
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_title(title,color="red")

plt.show()
```

```python
plots(T,P,"magic")
```
